Q2)a) and b)

```
import numpy as np
import builtins
import cmath
import matplotlib.pyplot as plt
points=[]
def To_create_coordinates(n):
  for j in range(n):
    points.append(np.array((np.random.uniform(-1,1),np.random.uniform(-1,1))))
val=int(input("Enter the no of coordinates (which should be greater than 15):"))
To_create_coordinates(val)
print("The coordinates of all the points are:",*points,sep="\n")
new_x=float(input("Enter the new x coordinate:"))
new_y=float(input("Enter the new y coordinate:"))
New_coordinate=np.array((new_x,new_y))
for j in range(val):
    plt.scatter(points[j][0],points[j][1],marker='o',color='blue')
points.sort(key=lambda x:np.linalg.norm(New_coordinate-x))
# print(points)
print("First 5 points nearest to the new coordinate:")
for j in range(5):
  print(points[j],sep="\n")
print("First 10 points nearest to the new coordinate:")
for j in range(10):
  print(points[j],sep="\n")
print("First 15 points nearest to the new coordinate:")
for j in range(15):
  print(points[j],sep="\n")

#For part (c) of the question to print the k positive angles
def To_printk_posAngles(k):
    c=0
    for j in range(val):
        z1=complex(points[j][0],points[j][1])
        z2=complex(new_x,new_y)
        if(c<k):
            if(np.angle(z1,deg=True)-np.angle(z2,deg=True)>0):
                print(points[j])
                plt.scatter(points[j][0],points[j][1],color='red')
                c+=1
        else:
            break
plt.scatter(new_x,new_y,color="green")
print("First 5 angles which are positive:")
To_printk_posAngles(5)
print("First 10 angles which are positive:")
To_printk_posAngles(10)
print("First 15 angles which are positive:")
To_printk_posAngles(15)

```

```
import numpy as np
import matplotlib.pyplot as plt
data_array=[]
sum,mean,f1,f2=[],[],[],[]
for i in range(100):
    data_array.append(np.random.uniform(-0.5, 0.5))
sum.append(data_array[0])
for i in range(1,100):
    sum.append(sum[i-1]+data_array[i])
for i in range(100):
    mean.append(sum[i]/(i+1))
    f1.append(1/(np.sqrt(i+1)))
    f2.append(-1/(np.sqrt(i+1)))
x=np.linspace(0,10,100)
plt.plot(x,f1,color='red')
plt.plot(x,f2,color="blue")
plt.plot(x,mean,color='green')
plt.show()
```
