OPTIMAL STRATERGY FOR FIRST BATTING:

```python
import pandas as pd

p_out_min=[0.01,0.02,0.03,0.1,0.3]
p_out_max=[0.1,0.2,0.3,0.5,0.7]
Action=[1,2,3,4,6]
p_out=[]
p_out.append([1,1,1,1,1])
for w in range(1,11):
    p_w=[]
    for a in range(5):
        p_w.append(p_out_max[a]+((p_out_min[a]-p_out_max[a])*(w-1)/9))
    p_out.append(p_w)
Bscore=[[0 for i in range(11)]for j in range(301)]
Bshot=[[0 for i in range(11)]for j in range(301)]
# Optimal Stratergy for batting first:
def optimal_stratergy(Bscore,Bshot,Action,p_out):
    b=1
    for w in range(1,11):
        for a in range(5):
            score=(1-p_out[w][a])*Action[a]
            if score>Bscore[1][w]:
                Bscore[1][w]=score
                Bshot[1][w]=Action[a]
    for b in range(2,301):
        for w in range(1,11):
            if b-w<=290:
                for a in range(5):
                    score=(1-p_out[w][a])*(Action[a]+Bscore[b-1][w])+(p_out[w][a]*Bscore[b-1][w-1])
                    if score>Bscore[b][w]:
                        Bscore[b][w]=score
                        Bshot[b][w]=Action[a]
    return Bscore,Bshot
Bscore,Bshot=optimal_stratergy(Bscore,Bshot,Action,p_out)
Best_score=pd.DataFrame(Bscore)
print("The Best score when b ball and w wicket left is:\n",Best_score[1:])
Best_shot=pd.DataFrame(Bshot)
print("The best shot when b ball and w wicket left is:\n",Best_shot[1:])
```
